```python
sns.countplot(data=df,x='Size',order=['Free','S','XS','M','L','XL','XXL','3XL','4XL','5XL','6XL'])
```

Groupby

```python
size_qty=df.groupby(['Size'],as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
sns.barplot(x='Size',y='Quantity',data=size_qty)
```

#### The highest amount of size sold were of M(Medium size)

Courier status

```python
# plt.figure(figsize=(10,20))
ax=sns.countplot(df,x='Courier Status',hue='Status')
for i in ax.containers:
    ax.bar_label(i)
```

#### Most of the orders were delivered to the buyer
